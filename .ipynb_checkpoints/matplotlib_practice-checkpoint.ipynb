```python
import matplotlib.pyplot as plt
```

```python
#list of variables to chart 
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan', 'Feb','Mar', 'Apri', 'May', 'Jun', ' Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Saving Account Results')
plt.show()
```

```python
#creating scatter plots using plt.plot()
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(y_axis, x_axis, 'o',c= 'red',label="Chicago")
plt.title('Pyber Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, y_axis_larger)
```

```python
#new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#object oriented scatter plot 
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=(y_axis), c="skyblue", label= "Boston", edgecolors="k", alpha=.8,)
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month', fontsize = 20)
ax.set_xlim(0,50)
ax.set_ylim("Dec", 'Jan')
ax.legend()
```

```python
#create pie charts
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
        explode=explode_values,
        labels = x_axis,
        autopct='%.1f%%',
        colors = colors)
plt.show()
```
